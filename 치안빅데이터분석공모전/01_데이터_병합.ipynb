## DATA3: NPA2020

```python
NPA2020
```

### 1) 교통사고 관련 데이터 추출

```python
idx4 = [i for i in range(len(NPA2020['EVT_CL_CD'])) if NPA2020['EVT_CL_CD'][i] in [4, 401, 402, 403, 404, 405, 406]]
NPA2020_4 = NPA2020.iloc[idx4, :]
NPA2020_4
```

```python
NPA2020_4.to_csv(path + 'NPA2020_4.csv', index=False)
```

### RECV_CPLT_TM6

```python
NPA2020_4 = pd.read_csv(path + 'NPA2020_4.csv')
```

```python
TM6 = []

for i in range(len(NPA2020_4['RECV_CPLT_TM'])):
  tm = str(NPA2020_4['RECV_CPLT_TM'][i])
  while True:
    if len(tm) == 6:
      TM6.append(tm)
      break
    tm = ''.join(('0', tm))

NPA2020_4['RECV_CPLT_TM6'] = TM6
```

### 2) 날짜 세분화

```python
# RECV_CPLT_DT(일자) datetime 형식으로 변경 

from datetime import datetime

DT_TM = []
datetime_format = "%Y%m%d"
for i in range(len(NPA2020_4['RECV_CPLT_TM6'])):
  datetime_string = str(NPA2020_4['RECV_CPLT_DT'][i])
  DT_TM.append(datetime.strptime(datetime_string, datetime_format))

NPA2020_4['RECV_CPLT_DT'] = DT_TM

NPA2020_4['RECV_CPLT_DT'] = NPA2020_4["RECV_CPLT_DT"].dt.strftime("%Y/%m/%d")
NPA2020_4
```

```python
NPA2020_4['YEAR'], NPA2020_4['MONTH'], NPA2020_4['DAY'] = NPA2020_4['RECV_CPLT_DT'].str.split('/', 2).str
NPA2020_4
```

```python
NPA2020_4['YEAR'].unique()
```

```python
NPA2020_4['YEAR'] = 20
NPA2020_4.head()
```

```python
# 교통 + 날짜 세분화 한 파일
NPA2020_4.to_csv(path + 'NPA2020_4_d.csv', index=False)
```

### 3) 시간 세분화

```python
NPA2020_4_d = pd.read_csv(path + 'NPA2020_4_d.csv')
NPA2020_4_d.head()
```

```python
# RECV_CPLT_TM 6자리로 변경 --> RECV_CPLT_TM6 열 추가

TM6 = []

for i in range(len(NPA2020_4_d['RECV_CPLT_TM'])):
  tm = str(NPA2020_4_d['RECV_CPLT_TM'][i])
  while True:
    if len(tm) == 6:
      TM6.append(tm)
      break
    tm = ''.join(('0', tm))

NPA2020_4_d['RECV_CPLT_TM6'] = TM6
```

```python
NPA2020_4_d.head()
```

```python
NPA2020_4_d['HOUR'] = NPA2020_4_d['RECV_CPLT_TM6'].str[:2]
NPA2020_4_d['MINUTE'] = NPA2020_4_d['RECV_CPLT_TM6'].str[2:4]
NPA2020_4_d['SECOND'] = NPA2020_4_d['RECV_CPLT_TM6'].str[4:]
NPA2020_4_d.head()
```

```python
# 교통 + 날짜 + 시간 세분화 저장
NPA2020_4_d.to_csv(path + 'NPA2020_4_d_t.csv', index=False)
```
