### 시간 데이터 범주화

```python
# 시간으로 새벽(0~4) | 아침(5~8) | 낮(9~16) | 저녁(17~20) | 밤(21~23) 범주화

hour_cat = []
for hour in data['HOUR']:
  if hour>=0 and hour<=4:
    hour_cat.append('DAWN')
  elif hour>=5 and hour<=8:
    hour_cat.append('MORN')
  elif hour>=9 and hour<=16:
    hour_cat.append('AFTE')
  elif hour>=17 and hour<=20:
    hour_cat.append('EVEN')
  elif hour>=21 and hour<=23:
    hour_cat.append('NIGH')

data['HOUR_CAT'] = hour_cat
data
```

```python
from datetime import datetime

# 날짜를 나타내는 DATE 열 추가
date = []
datetime_format = "%y%m%d"
for i in range(len(data['YEAR'])):
  datetime_string = ''.join((str(data['YEAR'][i]), str(data['MONTH'][i]), str(data['DAY'][i])))
  date.append(datetime.strptime(datetime_string, datetime_format).date())

data['DATE'] = date
data.head(3)
```

```python
from pytimekr import pytimekr

holi_20 = pytimekr.holidays(2020)  #2020년도 공휴일
holi_21 = pytimekr.holidays(2021)  #2021년도 공휴일
holi_22 = pytimekr.holidays(2022)  #2022년도 공휴일
holi_23 = pytimekr.holidays(2023)  #2023년도 공휴일
```

```python
holiday = []
for i in range(len(data['DATE'])):
  if data['YEAR'][i] == 20:
    if data['DATE'][i] in holi_20:
      holiday.append(1)
    else:
      holiday.append(0)

  if data['YEAR'][i] == 21:
    if data['DATE'][i] in holi_21:
      holiday.append(1)
    else:
      holiday.append(0)
  
  if data['YEAR'][i] == 22:
    if data['DATE'][i] in holi_22:
      holiday.append(1)
    else:
      holiday.append(0)

  if data['YEAR'][i] == 23:
    if data['DATE'][i] in holi_23:
      holiday.append(1)
    else:
      holiday.append(0)

data['HOLIDAY'] = holiday
data.head()
```

```python
import geopandas as gpd
from pyproj.crs import CRS

# 좌표 속성을 가지고 지오메트리 데이터를 생성
geom_df = gpd.points_from_xy(data_XY.HPPN_X, data_XY.HPPN_Y) # 405544

# 기존의 데이터프레임 컬럼명을 가져옴 
columns = data_XY.columns

# 속성 데이터프레임에 Geometry를 붙여서 공간 데이터프레임을 생성 
gdf_4326 = gpd.GeoDataFrame(data_XY[columns], geometry=geom_df, crs=CRS.from_epsg(4326)) # 405544

# 좌표계 변환 EPSG:4326(소수점 단위) --> EPSG:3857(미터 단위) 미터 단위로 격자 생성하기 위해
gdf_3857 = gdf_4326.to_crs('epsg:3857')

# # Geometry를 이용하여 X, Y 좌표 항목을 생성 -Pydeck 용
# gdf_KP2020_4326['lon'] = gdf_KP2020_4326.geometry.x
# gdf_KP2020_4326['lat'] = gdf_KP2020_4326.geometry.y
```

```python
from shapely.geometry import Polygon
xmin,ymin,xmax,ymax = gdf_3857.total_bounds

# 가로 세로 500m 크기의 격자 생성
length = 500  #세로 길이
wide = 500  #가로 길이

cols = list(range(int(np.floor(xmin)), int(np.ceil(xmax)), wide))
rows = list(range(int(np.floor(ymin)), int(np.ceil(ymax)), length))
rows.reverse()

polygons = []
for x in cols:
  for y in rows:
    polygons.append( Polygon([(x,y), (x+wide, y), (x+wide, y-length), (x, y-length)]) )
grid500 = gpd.GeoDataFrame({'geometry':polygons})
grid500.crs = "epsg:3857"
grid500_4326 = grid500.to_crs('epsg:4326')   #epsg:4326으로 좌표 변환
grid500_4326['index'] = grid500_4326.index  #폴리곤별 인덱스를 컬럼으로 생성 - 나중에 속성조인을 위해
```

```python
grid500_4326
```

```python
grid_data = gpd.sjoin(grid500_4326, gdf_4326)   #기존데이터와 폴리곤데이터 결합
```

```python
# 인구데이터 격자별 (인구수/중심좌표수) 구한 열 추가
pop_df = pop_df[pop_df['CENT_CNT']!=0].reset_index(drop=True)
pop_df = pop_df.fillna(0)
pop_df['pop/cent'] = pop_df['val']/pop_df['CENT_CNT']
```

```python
# 시간 오래 걸림(약 8분 소요)
# 만든 격자에 인구수 할당 
pop = [0] * len(grid500['geometry'])  #각 격자별 인구수를 저장할 리스트. 0으로 초기화

for i in range(len(grid500['centroid'])):
  df = pd.DataFrame({'tf':grid500['centroid'][i].within(pop_df['geometry'])})
  df_true = df[df['tf']==True]
  if len(df_true) != 0:
    pop_sum = 0
    for j in range(len(df_true)):
      pop_sum += pop_df['pop/cent'][df_true.index[j]]
    pop[i] = pop_sum

grid500.insert(3, 'POPULATION', pop)
```
